```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import random
import string
```

### Method Section

Just trust these methods...

```python
def generate_alphanumeric(str_len):
    return ''.join(
        random.choices(string.ascii_letters + string.digits, k=str_len))
```

```python
def scrape_by_id(tid, directory, folder_guide):
    imdb = "https://imdb.com/"
    
    url = imdb + "title/"+ tid +"/reviews?ref_=tt_ql_3"
    r  = requests.get(url)
    html = r.text

    soup = BeautifulSoup(html, features='html.parser')
    lister = soup.findAll("div", {"class": "lister-item-content"})

    max_review = 10
    c = 0
    for i in range(len(lister)):
        l = lister[i]

        if not l.find('span', {'class' : 'spoiler-warning'}):
            try :
                rate_node = str(l.findChildren('span', {'class' : 'rating-other-user-rating'})[0].find('span'))
                rate = ''.join(list(rate_node))[6:-7]
                a = l.findChildren("a" , recursive=False)[0]
                path = a['href']
                r_path = requests.get(imdb + path)
                r_html = r_path.text
                review_node = BeautifulSoup(r_html, features='html.parser').find('div', {'class' : 'text show-more__control'})
                review = ''.join(list(str(review_node))[37:-6])
                review = re.sub('<br/>', '', review)
                o = open(directory +'/'+folder_guide[rate]+'/'+tid+'_'+generate_alphanumeric(10)+'_'+rate+'.txt', 'w')
                o.write(review)
                o.close()
                print('Success: ', rate)
                c += 1
                if c >= max_review:
                    break
            except :
                print ('gagal')
```

### Read the tid

```python
s = pd.read_csv('movie2010en.txt', names=['tid'])
```

### Folder Mapping (Ingat 1-2 d gabung dan yada2)

```python
folder_map = dict()
for i in range(5):
    folder_name = str(2*i + 1) + ' to ' + str(2*(i+1))
    folder_map[str(2*i + 1)] = folder_name
    folder_map[str(2*(i+1))] = folder_name
folder_map
```

```python
for tid in s.iloc[0:100]['tid'].values:
    scrape_by_id(tid, 'final dataset', folder_map)
```

```python
def fetch_data(dataframe, start, finish, directory, folder_guide):
    for tid in dataframe.iloc[start:finish]['tid'].values:
        scrape_by_id(tid, 'dataset', folder_guide)
```
