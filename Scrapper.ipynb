```python
def scrape_by_id(tid):
    imdb = "https://imdb.com/"
    
    url = imdb + "title/"+ tid +"/reviews?ref_=tt_ql_3"
    r  = requests.get(url)
    html = r.text

    soup = BeautifulSoup(html, features='html.parser')
    lister = soup.findAll("div", {"class": "lister-item-content"})

    max_review = 10
    c = 0
    for i in range(len(lister)):
        l = lister[i]

        if not l.find('span', {'class' : 'spoiler-warning'}):
            try :
                rate_node = str(l.findChildren('span', {'class' : 'rating-other-user-rating'})[0].find('span'))
                rate = ''.join(list(rate_node))[6:-7]

                a = l.findChildren("a" , recursive=False)[0]
                path = a['href']
                r_path = requests.get(imdb + path)
                r_html = r_path.text
                review_node = BeautifulSoup(r_html, features='html.parser').find('div', {'class' : 'text show-more__control'})
                review = ''.join(list(str(review_node))[37:-6])
                review = re.sub('<br/>', '', review)

                o = open(rate+'/'+tid+'.txt', 'w')
                o.write(review)
                o.close()

                c += 1
                if c >= max_review:
                    break
            except :
                print ('gagal')
```
